```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

def gender_label(year, sp):
    is_year = women_degrees['Year'] == year
    return label_position(100-women_degrees[is_year].iloc[0][sp], women_degrees[is_year].iloc[0][sp])
    
def label_position(men, women):
    if abs(men - women) > 10:
        if men > women:
            men = men + 10
            women = women - 10
        else:
            men = men - 10
            women = women + 10
        return men, women
    else:
        return None

def plot_categories(category_number, category):
    for i, sp in enumerate(category):
        ax = fig.add_subplot(6, 3, (3*i)+category_number)
        ax.plot(women_degrees['Year'], women_degrees[sp], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[sp], c=cb_orange, label='Men', linewidth=3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0, 100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        ax.set_title(sp)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
        
        if i == 0 or i == len(category)-1:
            men_pos, women_pos = gender_label(2005,sp) or (None, None)
            if men_pos is not None:
                ax.text(2005, men_pos, 'Men')
                ax.text(2005, women_pos, 'Women')
            if i == len(category) - 1:
                ax.tick_params(labelbottom="on")

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(18, 18))

categories = [stem_cats, lib_arts_cats, other_cats]

for i, category in enumerate(categories):
    plot_categories(i+1, category)

plt.savefig('gender_degrees.png')
plt.show()
```
